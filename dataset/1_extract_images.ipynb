```python
from pathlib import Path
from huggingface_hub import hf_hub_download
from typstscribe.dataset.common import HUANG_DATASET_REPO_ID, TEMP_DIR_OUTSIDE_REPO
```

```python
TEMP_DIR_OUTSIDE_REPO.mkdir(parents=True, exist_ok=True)
```

```python
# Download images
images_rar_path = Path(hf_hub_download(
    repo_id=HUANG_DATASET_REPO_ID,
    filename="root.rar",
    repo_type="dataset"
)).resolve()

assert images_rar_path.exists()
images_rar_path
```

```python
# Create temporary folder with extracted images
# Working with the rarfile directory in Python turned out to be a massive pain in the proverbial arse
# on both Linux and Windows.
# To work around this, we unrar into a temporary directory and work with the filesystem when we create the dataset
image_rar_path_str = str(images_rar_path)
temporary_image_folder_str = str(TEMP_DIR_OUTSIDE_REPO)
! unrar x -o+ "$image_rar_path_str" "$temporary_image_folder_str" | awk 'NR % 50000 == 0'
```
