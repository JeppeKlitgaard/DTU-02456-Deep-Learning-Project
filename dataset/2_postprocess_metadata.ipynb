```python
import pandas as pd
from common import RAW_METADATA_PARQUET_FILE_PATH, METADATA_PARQUET_FILE_PATH
```

```python
assert RAW_METADATA_PARQUET_FILE_PATH.exists(), f"{RAW_METADATA_PARQUET_FILE_PATH} does not exist"
assert METADATA_PARQUET_FILE_PATH.parent.exists(), f"{METADATA_PARQUET_FILE_PATH.parent} does not exist"
```

```python
df_raw = pd.read_parquet(RAW_METADATA_PARQUET_FILE_PATH)
df_raw
```

```python
# The Hoang dataset has duplicates, also across splits, which we definitely want to avoid like the plague
df_dedup = df_raw.drop_duplicates(subset=["image_filename"], keep="first").reset_index(drop=True)
```

```python
df_dedup.to_parquet(METADATA_PARQUET_FILE_PATH, index=False)
```
