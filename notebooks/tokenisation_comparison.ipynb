```python
from transformers import TrOCRProcessor, AutoTokenizer
from typstscribe.const import PROJECT_DIR, MODELS_DIR
import json

TOKENIZER_MODELS_DIR = MODELS_DIR / "tokenizers"
REPORT_DIR = PROJECT_DIR / "Documents" / "Report"
```

```python
# TYPST_TOKENIZER_PATH = "JeppeKlitgaard/typst-tokenizer"
TYPST_TOKENIZER_PATH = TOKENIZER_MODELS_DIR / "typst_tokenizer"
TROCR_MODEL_PATH = "microsoft/trocr-small-stage1"

typst_tokenizer = AutoTokenizer.from_pretrained(TYPST_TOKENIZER_PATH)
default_tokenizer = TrOCRProcessor.from_pretrained(TROCR_MODEL_PATH, use_fast=False).tokenizer
```

```python
# SAMPLE_TEXT = (REPORT_DIR / "typst_sample.typ").read_text()
# SAMPLE_TEXT = SAMPLE_TEXT.removeprefix("$\n").removesuffix("\n$\n")

SAMPLE_TEXT = """underbrace(f(theta), "obj")"""
```

```python

display(typst_tokenizer.tokenize(SAMPLE_TEXT))
display(len(typst_tokenizer.tokenize(SAMPLE_TEXT)))
```

```python

display(default_tokenizer.tokenize(SAMPLE_TEXT))
display(len(default_tokenizer.tokenize(SAMPLE_TEXT)))
```

```python
print(default_tokenizer.decode(default_tokenizer.encode(SAMPLE_TEXT)))
print(typst_tokenizer.decode(typst_tokenizer.encode(SAMPLE_TEXT)))
```

```python
(REPORT_DIR / "typst_tokenizer_example.json").write_text(json.dumps(typst_tokenizer.tokenize(SAMPLE_TEXT)))

typst_example = typst_tokenizer.tokenize(SAMPLE_TEXT)
typst_example = list(map(lambda x: x.replace("Ġ", " "), typst_example))
(REPORT_DIR / "typst_tokenizer_example.json").write_text(json.dumps(typst_example))

default_example = default_tokenizer.tokenize(SAMPLE_TEXT)
default_example = list(map(lambda x: x.replace("\u2581", " "), default_example))
(REPORT_DIR / "default_tokenizer_example.json").write_text(json.dumps(default_example))
```
